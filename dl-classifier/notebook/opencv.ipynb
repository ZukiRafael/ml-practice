```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
data_loc = '../data/train/NORMAL/IM-0119-0001.jpeg'
img = cv.imread(data_loc, 0)
```

```python
plt.imshow(img,
               cmap='gray_r',
               vmin=0,
               vmax=255)
```

```python
img_size = (100,100)
img_resize = cv.resize(img, img_size, interpolation=cv.INTER_AREA)
plt.imshow(img_resize,
               cmap='gray_r',
               vmin=0,
               vmax=255)
```
